```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Morocco.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.heatmap(df.isnull(), cmap="viridis")
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
irrelevant_columns = ['description', 'severerisk', 'preciptype', 'sunrise', 'sunset', 'stations']
df.drop(columns=irrelevant_columns, inplace=True)
df.rename(columns={'name': 'item_id', "datetime" : "timestamp"}, inplace=True)
```

```python
df = pd.get_dummies(df, columns=['conditions', 'icon'])
```

```python
sns.set_theme(style='whitegrid')
plt.figure(figsize=(20, 20))

num_cols = len(df.columns)
for i, col in enumerate(df.columns, start=1):
    plt.subplot(num_cols, 1, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

df.iloc[:, 2:] = df.iloc[:, 2:].astype(float)

scaler = StandardScaler()
df.iloc[:, 2:] = scaler.fit_transform(df.iloc[:, 2:])
```

```python
corr_matrix = df.iloc[:, 1:].corr()
sns.heatmap(corr_matrix, cmap="viridis")
```

```python
threshold = 0.8
mask = np.abs(corr_matrix) > threshold

indices = np.where(mask)

for row, col in zip(indices[0], indices[1]):
    if row != col and row < col:
        print(f"Correlation between '{corr_matrix.index[row]}' and '{corr_matrix.columns[col]}' is {corr_matrix.iloc[row, col]}")
```

```python
correlated_cols = ["feelslikemax", "feelslikemin", "feelslike", "icon_rain", "solarenergy", "uvindex"]
for i in df.columns:
    if "conditions" in i:
        correlated_cols.append(i)
df.drop(columns=correlated_cols, inplace=True)
```

```python
corr_matrix = df.iloc[:, 1:].corr()
sns.heatmap(corr_matrix, cmap="viridis")
threshold = 0.7
mask = np.abs(corr_matrix) > threshold

indices = np.where(mask)

for row, col in zip(indices[0], indices[1]):
    if row != col and row < col:
        print(f"Correlation between '{corr_matrix.index[row]}' and '{corr_matrix.columns[col]}' is {corr_matrix.iloc[row, col]}")
```
